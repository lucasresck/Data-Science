# Decision trees

Decision trees are supervised machine learning algorithms based on trees. They try to learn how to classify or map the data based on a train dataset. At the end, we have a tree of decisions which we can use to predict new data.

**Goals**:
- Implement some decision trees with *scitkit-learn*;
- Train decision trees;
- Predict with decision trees.

First, a simple example, with 2 points:

```python
from sklearn import tree
from sklearn.datasets.samples_generator import make_blobs
```

```python
# Data
X = [[0, 0], [1, 1]]
Y = [0, 1]

# Instantiate, train and plot of a decision tree
clf = tree.DecisionTreeClassifier()
clf.fit(X, Y)
tree.plot_tree(clf);
```

When we predict a point, it seems reasonable.

```python
print(clf.predict([[0.9, 0.9]]))  # Predict
print(clf.predict_proba([[0.9, 0.9]]))  # Probabilities
```
